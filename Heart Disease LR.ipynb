```python
import pandas as pd

df = pd.read_csv('heart_attack_youngsters_india.csv')
display(df.head())
```

```python
print("Shape of the DataFrame:")
print(df.shape)

print("\nData types of each column:")
print(df.dtypes)

print("\nDescriptive statistics for numerical columns:")
display(df.describe())

print("\nUnique value counts for categorical columns:")
for col in df.select_dtypes(include='object').columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())

print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
categorical_cols = ['Gender', 'Region', 'Urban/Rural', 'SES', 'Smoking Status', 'Exercise Induced Angina', 'Heart Attack Likelihood']
for col in categorical_cols:
    print(f"Unique values in {col}: {df[col].unique()}")

print(f"\nUnique values in Blood Pressure (first 50): {df['Blood Pressure (systolic/diastolic mmHg)'].unique()[:50]}")
```

```python
import numpy as np

numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Column '{col}': Found {len(outliers)} outliers.")

    # Capping outliers
    df[col] = np.where(
        df[col] < lower_bound,
        lower_bound,
        np.where(df[col] > upper_bound, upper_bound, df[col])
    )
    print(f"Column '{col}': Capped outliers.")
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
print(f"Numerical columns identified: {numerical_cols}")
```

```python
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

display(df.head())
```

```python
y = df['Heart Attack Likelihood']
X = df.drop('Heart Attack Likelihood', axis=1)

categorical_cols = X.select_dtypes(include='object').columns
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

display(X_encoded.head())
```

```python
from imblearn.over_sampling import SMOTE

print("Original target variable distribution:")
print(y.value_counts())

# Check for imbalance
if y.value_counts()['Yes'] / y.value_counts()['No'] < 0.5: # Example threshold for imbalance
    print("\nClass imbalance detected. Applying SMOTE...")
    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X_encoded, y)

    print("\nResampled target variable distribution:")
    print(y_resampled.value_counts())
else:
    print("\nNo significant class imbalance detected.")
    X_resampled = X_encoded
    y_resampled = y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.20, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100], 'penalty': ['l2']}

# Instantiate Logistic Regression model
logreg = LogisticRegression(solver='liblinear', max_iter=1000) # Using liblinear solver for smaller datasets and l2 penalty

# Instantiate GridSearchCV
grid_search = GridSearchCV(logreg, param_grid, cv=5, scoring='roc_auc')

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Get the best model
best_model = grid_search.best_estimator_

# Predict on the test set
y_pred = best_model.predict(X_test)
y_prob = best_model.predict_proba(X_test)[:, 1] # Probability of the positive class

# Calculate and print confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Calculate and print classification report
class_report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(class_report)

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob, pos_label='Yes')

# Calculate AUC
auc = roc_auc_score(y_test, y_prob)
print(f"\nROC AUC Score: {auc:.4f}")

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

# Train a Logistic Regression model with default hyperparameters
logreg_default = LogisticRegression(random_state=42, solver='liblinear', max_iter=1000)
logreg_default.fit(X_train, y_train)

# Predict probabilities on the test set
y_prob_default = logreg_default.predict_proba(X_test)[:, 1]

# Calculate the ROC AUC score for the default model
auc_default = roc_auc_score(y_test, y_prob_default)
print(f"ROC AUC Score (Default Model): {auc_default:.4f}")

# Compare with the GridSearchCV-tuned model
print(f"ROC AUC Score (Tuned Model): {auc:.4f}")

if auc > auc_default:
    print("\nHyperparameter tuning improved the ROC AUC score.")
elif auc < auc_default:
    print("\nHyperparameter tuning slightly decreased the ROC AUC score.")
else:
    print("\nHyperparameter tuning did not change the ROC AUC score significantly.")
```

```python
# Get the best trained Logistic Regression model
best_model = grid_search.best_estimator_

# Access the coefficients
coefficients = best_model.coef_[0]

# Get feature names
feature_names = X_encoded.columns

# Create a Series to associate feature names with coefficients
coef_series = pd.Series(coefficients, index=feature_names)

# Sort coefficients by absolute value in descending order
sorted_coef = coef_series.abs().sort_values(ascending=False)

# Print the top and bottom coefficients (e.g., top 10 and bottom 10)
print("Top 10 most important features (based on absolute coefficient value):")
display(coef_series[sorted_coef.index[:10]])

print("\nBottom 10 most important features (based on absolute coefficient value):")
display(coef_series[sorted_coef.index[-10:]])
```

```python
# Print best hyperparameters
print("Best Hyperparameters:", grid_search.best_params_)

# Parse classification report string to get metrics
class_report_dict = classification_report(y_test, y_pred, output_dict=True)

accuracy = class_report_dict['accuracy']
precision = class_report_dict['weighted avg']['precision']
recall = class_report_dict['weighted avg']['recall']
f1_score = class_report_dict['weighted avg']['f1-score']

print(f"\nModel Accuracy: {accuracy:.4f}")
print(f"Model Precision: {precision:.4f}")
print(f"Model Recall: {recall:.4f}")
print(f"Model F1-score: {f1_score:.4f}")

# Print ROC-AUC
print(f"\nROC-AUC: {auc:.4f}")

# Print top 10 important feature weights
print("\nTop 10 Most Important Feature Weights:")
display(coef_series[sorted_coef.index[:10]])
```
